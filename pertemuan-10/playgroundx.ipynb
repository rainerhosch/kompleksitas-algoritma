```python
def tambahSampaiKe(n):
    jumlah = 0
    for i in range(n, 0, -1):
        jumlah += 1
    return jumlah
```

```python
#TSP
import itertools

# Representasi jarak antar kota (matrix)
distances = [
    [0, 10, 15, 20],
    [10, 0, 35, 25],
    [15, 35, 0, 30],
    [20, 25, 30, 0]
]

n = len(distances)
cities = range(n)

min_path = None
min_cost = float('inf')

for perm in itertools.permutations(cities[1:]):  # kota 0 sebagai start
    path = (0,) + perm + (0,)
    cost = sum(distances[path[i]][path[i+1]] for i in range(len(path)-1))
    if cost < min_cost:
        min_cost = cost
        min_path = path

print("Rute terbaik:", min_path)
print("Biaya minimum:", min_cost)
```

```python
#Knapsack
import itertools

items = [
    {"name": "A", "weight": 3, "value": 60},
    {"name": "B", "weight": 2, "value": 100},
    {"name": "C", "weight": 4, "value": 120}
]

capacity = 5
best_value = 0
best_combination = None

for r in range(len(items)+1):
    for combo in itertools.combinations(items, r):
        total_weight = sum(item["weight"] for item in combo)
        total_value = sum(item["value"] for item in combo)
        if total_weight <= capacity and total_value > best_value:
            best_value = total_value
            best_combination = combo

print("Kombinasi terbaik:", [item["name"] for item in best_combination])
print("Nilai maksimum:", best_value)
```

```python
#String Matching dengan Brute Force

def brute_force_string_match(text, pattern):
    n = len(text)
    m = len(pattern)
    for i in range(n - m + 1):
        match = True
        for j in range(m):
            if text[i + j] != pattern[j]:
                match = False
                break
        if match:
            return i  # posisi ditemukan
    return -1  # tidak ditemukan

text = "abracadabra"
pattern = "cad"

pos = brute_force_string_match(text, pattern)
if pos != -1:
    print(f"Pattern '{pattern}' ditemukan pada posisi {pos}")
else:
    print(f"Pattern '{pattern}' tidak ditemukan")
```

```python
# Contoh algoritma brute force: mencari pasangan dua angka dalam list yang jumlahnya sama dengan target

def pasanganJumlahTarget(nums, target):
    """
    Strategi brute force untuk mencari semua pasangan (i, j) dalam daftar nums
    yang jika dijumlahkan hasilnya sama dengan target.
    """
    hasil = []
    n = len(nums)
    for i in range(n):
        for j in range(i+1, n):  # memastikan i < j agar tidak duplikat dan tidak menjumlahkan diri sendiri
            if nums[i] + nums[j] == target:
                hasil.append((nums[i], nums[j]))
    return hasil

# Contoh penggunaan
nums = [2, 7, 11, 15, 1, 8]
target = 9
print(pasanganJumlahTarget(nums, target))
```
